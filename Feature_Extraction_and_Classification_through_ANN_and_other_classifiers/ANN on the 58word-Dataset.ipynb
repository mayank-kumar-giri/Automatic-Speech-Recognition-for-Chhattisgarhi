 # Implementing Regional Language(Chhattisgarhi) ASR using ANN

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading previously extracted features of the 2320 audio files.
x_train = np.load("x_train.npy")
y_train = np.load("y_train.npy")
x_test = np.load("x_test.npy")
y_test = np.load("y_test.npy")
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping

model = Sequential()
model.add(Dense(193,input_dim=193, activation='relu'))
model.add(Dense(106))

model.add(Dense(58, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=500, batch_size=500)

#Evaluating on x_test and printing the 232 predictions
scores = model.evaluate(x_test, y_test)
val_loss,val_acc = model.evaluate(x_test, y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
print(scores)
print(val_acc,val_loss)
classes = model.predict([x_test])
pred = []
for i in classes:
    pred.append(np.argmax(i))
    print(np.argmax(i),end=' ')
print(len(y_test),len(pred))
```

```python
import time
start = time.perf_counter()
acc = []
#Training multiple models on variable epochs. Saving the ones having Accuracy>38%
for i in range(5,501,5):
    model.fit(x_train, y_train, epochs=i, batch_size=500)
    val_loss,val_acc = model.evaluate(x_test, y_test)
    acc.append(val_acc*100)
    temp = val_acc*100
    if temp>=38:
        model.save("Model_Accuracy_"+str(temp))
xx = list(range(5,501,5))
plt.plot(xx,acc)
plt.show()
print("Time taken - ",time.perf_counter()-start)
```

```python
#Printing the maximum Accuracy
ind = acc.index(max(acc))
print(acc[ind],xx[ind])
print(len(acc))
print(len(xx))
```

```python
classes = model.predict([x_test])
predicted_labels = []
for i in classes:
    predicted_labels.append(np.argmax(i))
print(y_test,predicted_labels)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predicted_labels)
accuracy = (accuracy*100)
```

```python
num = [0]*58
den = [0]*58
for i in range(len(predicted_labels)):
    den[predicted_labels[i]] += 1
    if predicted_labels[i]==y_test[i]:
        num[predicted_labels[i]] += 1
precisions = [0]*58
print("Precisions word-wise:")
for i in range(58):
    try:
        precisions[i]=(num[i]/den[i])*100
    except:
        precisions[i] = -1
        pass
    print(hin_dict[i],precisions[i])
```

```python
print("ACC",accuracy)
print("PRE",precisions)
# -1 indicates (0/0), i.e., no sample was classified to that class.
```
