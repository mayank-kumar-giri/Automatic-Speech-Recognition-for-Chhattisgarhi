```python
from os import listdir
from os.path import isfile, join
import os
import librosa
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import specgram
import json
import time
import io
from sklearn.ensemble import RandomForestClassifier
%matplotlib inline
plt.style.use('ggplot')
```

```python
x_train = np.load("x_train.npy")
y_train = np.load("y_train.npy")
x_test = np.load("x_test.npy")
y_test = np.load("y_test.npy")
```

```python
model = RandomForestClassifier(n_estimators=4600,criterion="entropy")
print("Training model...")
start = time.perf_counter()
#train model
model.fit(x_train,y_train)
print("Time taken for training -",time.perf_counter() - start)
start = time.perf_counter()
predicted_labels = model.predict(x_test)
print("Time taken for testing -",time.perf_counter() - start)
# print(y_test,predicted_labels)
```

```python
count = 0
for i in range(len(predicted_labels)):
    if predicted_labels[i]==y_test[i]:
        count+=1
acc = count/len(predicted_labels)
acc = acc*100
print("The accuracy is",acc,"%.")
```

```python
with open('eng_dict.json') as data_file:
    eng_dict = json.load(data_file)

from sklearn.externals import joblib
filename = 'hin_dict'
hin_dict = joblib.load(filename)
```

```python
num = [0]*58
den = [0]*58
for i in range(len(predicted_labels)):
    den[predicted_labels[i]] += 1
    if predicted_labels[i]==y_test[i]:
        num[predicted_labels[i]] += 1
precisions = [0]*58
print("Precisions word-wise:")
for i in range(58):
    try:
        precisions[i]=(num[i]/den[i])*100
    except:
        precisions[i] = -1
        pass
    print(hin_dict[i],precisions[i])
```

```python
print("ACC",acc)
print("PRE",precisions)
```

# Plotting Accuracy vs No. of Trees for best RF performance

```python
accuracies153 = []
xforaccuracies = list(range(1000,5000,50))
for i in range(1000,5000,50):
    model = RandomForestClassifier(n_estimators=1000,criterion="entropy")
    print("Training model...")
    start = time.perf_counter()
    #train model
    model.fit(x_train,y_train)
    print("Time taken for training -",time.perf_counter() - start)
    start = time.perf_counter()
    predicted_labels = model.predict(x_test)
    print("Time taken for testing -",time.perf_counter() - start)
    # print(y_test,predicted_labels)
    count = 0
    for i in range(len(predicted_labels)):
        if predicted_labels[i]==y_test[i]:
            count+=1
    acc = count/len(predicted_labels)
    acc = acc*100
    accuracies153.append(acc)
    print("The accuracy is",acc,"%.")
```

```python
print(accuracies153)
```

```python
plt.plot(xforaccuracies,accuracies153)
plt.show()
print(xforaccuracies[accuracies153.index(max(accuracies153))])
```
