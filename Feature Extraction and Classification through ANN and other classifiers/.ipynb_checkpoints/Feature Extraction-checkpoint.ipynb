# Feature extraction using the Five mentioned Techniques

```python
from os import listdir
from os.path import isfile, join
import os
import librosa
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import specgram
import json
import time
import io
from sklearn.ensemble import RandomForestClassifier
%matplotlib inline
plt.style.use('ggplot')
```

```python
with open('eng_dict.json') as data_file:
    eng_dict = json.load(data_file)
```

```python
dirname = os.getcwd()
train_folder = dirname + "\Dataset_Final\Training_Dataset"
test_folder = dirname + "\Dataset_Final\Testing_Dataset"
print(train_folder)
# make dict - index:(eng,hin)
# folders = [f for f in listdir(train_folder) if not isfile(join(train_folder, f))]
# label_eng_dict
```

```python
def extract_feature(file_name):
    X, sample_rate = librosa.load(file_name)
#     print("Features :",len(X), "sampled at ", sample_rate, "hz")
    stft = np.abs(librosa.stft(X))
    mfccs = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40).T,axis=0)
    chroma = np.mean(librosa.feature.chroma_stft(S=stft, sr=sample_rate).T,axis=0)
    mel = np.mean(librosa.feature.melspectrogram(X, sr=sample_rate).T,axis=0)
    contrast = np.mean(librosa.feature.spectral_contrast(S=stft, sr=sample_rate).T,axis=0)
    tonnetz = np.mean(librosa.feature.tonnetz(y=librosa.effects.harmonic(X), sr=sample_rate).T,axis=0)
    return mfccs,chroma,mel,contrast,tonnetz

def parse_audio_files(dirname):
    features, labels = np.empty((0,193)), np.empty(0)    
    print("dir",dirname)
    folders = [f for f in listdir(dirname) if not isfile(join(dirname, f))]
    for i in folders:
        foldname = dirname + "\\" + i

        onlyfiles = [f for f in listdir(foldname) if isfile(join(foldname, f))]

        foldname = foldname + "\\"
        print("\nFolder -",foldname,onlyfiles,"\n")
        for j in onlyfiles:
            temp = foldname
            temp = temp + j
            
            try:
                mfccs, chroma, mel, contrast, tonnetz = extract_feature(temp)
                ext_features = np.hstack([mfccs,chroma,mel,contrast,tonnetz])
                features = np.vstack([features,ext_features])
                labels = np.append(labels, eng_dict[j.split("_")[0]])
#                 print("File -",j)
            except:
                print("Error processing " + j + " - skipping")
    return np.array(features), np.array(labels, dtype = np.int)
```

```python
start = time.perf_counter()
x, y = parse_audio_files(train_folder)
test_x, test_y = parse_audio_files(test_folder)
print(x,y)
print("Time taken for Feature Extraction : ",time.perf_counter()-start)
```
